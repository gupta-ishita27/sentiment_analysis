```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
#reading reviews
dataset = pd.read_csv("reviews.csv")
dataset["star"] = dataset["Stars"].str[:3]
#dataset
```

```python
#bar plot on star ratings
summarised_results = dataset["star"].value_counts()
plt.bar(summarised_results.keys(), summarised_results.values)
plt.show()
```

```python
comment_words = '' 
stopwords = set(STOPWORDS) 

  
# iterate through the csv file 
for val in dataset["Comment"]: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
  
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower()
        #print(tokens[i]) 
        comment_words += " ".join(tokens)+" "
print(comment_words)

wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10,
                max_words = 100).generate(comment_words)


```

```python
# plot the WordCloud image                        
plt.figure(figsize = (10, 10), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
   
```
